## Erosion:-

OpenCV-Python is a library of Python bindings designed to solve computer vision problems. cv2.erode() method is used to perform erosion on the image. The basic idea of erosion is just like soil erosion only, it erodes away the boundaries of a foreground object (Always try to keep foreground in white). It is normally performed on binary images. It needs two inputs, one is our original image, the second one is called the structuring element or kernel which decides the nature of operation. A pixel in the original image (either 1 or 0) will be considered 1 only if all the pixels under the kernel are 1, otherwise it is eroded (made to zero).

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

image = cv2.imread('img_1.png')
rgb_image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
_kernel = np.ones((5, 5), np.uint8)
constant_image = cv2.erode(rgb_image, _kernel)
reflected_image = cv2.erode(rgb_image, _kernel, borderType=cv2.BORDER_REFLECT)

count = 3
titles = ['Original Image', 'Eroded Image', 'Reflected Image']
images = [rgb_image, reflected_image, constant_image]

for i in range(count):
    plt.subplot(1, 3, i + 1)
    plt.imshow(images[i])
    plt.title(titles[i])
    plt.axis('off')
plt.show()



```
