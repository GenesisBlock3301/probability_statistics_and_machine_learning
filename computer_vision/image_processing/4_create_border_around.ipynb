## Python OpenCV | cv2.copyMakeBorder() method

OpenCV-Python is a library of Python bindings designed to solve computer vision problems. cv2.copyMakeBorder() method is used to create a border around the image like a photo frame.

**Syntax:** cv2.copyMakeBorder(src, top, bottom, left, right, borderType, value
**src** --> image src.<br>
**top** --> width in number of pixel in top direction.<br>
**bottom** --> border width in number of pixel in bottom direction.<br>
**left** --> border width in number of pixel in left direction.<br>
**right** --> border with in number of pixel in right direction.<br>
**borderType** --> depict what kind of border to be added, e.g, cv2.BORDER_CONSTANT, cv2.BORDER_REFLECT.<br>
**dest** --> destination of image.<br>
**value** --> border color.<br>
Return an image


```python
import cv2
import matplotlib.pyplot as plt

image = cv2.imread('img_1.png')
rgb_image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

border_image = cv2.copyMakeBorder(rgb_image, 10, 10, 10, 10, cv2.BORDER_CONSTANT)

plt.subplot(121), plt.imshow(rgb_image), plt.title('Original')
plt.subplot(122), plt.imshow(border_image), plt.title('Border')
plt.show()
```
