# Interpolation Method:
Interpolation methods for image resizing are mathematical techniques used to estimate new pixel values when resizing an image-- especially when scaling up(enlarge)

## Choice of Interpolation Method for Resizing:
- cv2.INTER_AREA: This is used when we need to shrink an image.
- cv2.INTER_CUBIC: This is slower but more efficient.
- cv2.INTER_LINEAR: This is primarily used when zooming is required. This is the default interpolation technique in OpenCV.

#### Syntax: cv2.resize(source, dsize, dest, fx, fy, interpolation)
- source → Source of image
- dsize --> size of output array
- dest --> output array
- fx → scale factor along the horizontal axis X
- fy → scale factor along the horizontal axis Y

```python
import cv2
import matplotlib.pyplot as plt

image = cv2.imread('img.png')
rgb_image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

half = cv2.resize(rgb_image, (0, 0), fx=0.5, fy=0.5)
bigger = cv2.resize(rgb_image, (1050, 1610))
stretch_near = cv2.resize(rgb_image, (400, 400), interpolation=cv2.INTER_LINEAR)

titles = ['Original', 'Half', 'Bigger', 'Stretch Near']
images = [rgb_image, half, bigger, stretch_near]
count = 4
print(rgb_image.shape)
for i in range(count):
    plt.subplot(2, 2, i + 1)
    plt.imshow(images[i], 'gray')
    plt.title(titles[i])
plt.show()
```
