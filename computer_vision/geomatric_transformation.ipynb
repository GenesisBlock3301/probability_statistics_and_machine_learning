```python
import cv2
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('img.png')
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.imshow(gray, cmap='gray')
plt.axis('off')
plt.show()
```

```python
# resize -> provide specific width and height
new_width, new_height = 400, 300
resize_image = cv2.resize(image, (new_width, new_height))
plt.imshow(resize_image)
plt.axis('off')
plt.show()
resize_image.shape
```

```python
# rescaling -> provide percentage of size.
res = cv2.resize(image,None, fx=0.5, fy=0.5, interpolation=cv2.INTER_CUBIC)
plt.imshow(res)
# plt.axis('off')
# plt.xticks([0, 50, 100, 150, 200])  # Set specific X-axis ticks
# plt.yticks([0, 50, 100, 150, 200])
plt.show()
res.shape
```
