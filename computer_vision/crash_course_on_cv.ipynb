```python
import cv2
import matplotlib.pyplot as plt

from probability.standard_deviation import weight_column
```

```python
# Load the image
image = cv2.imread('img.png')
rgb_image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
gray = cv2.cvtColor(rgb_image, cv2.COLOR_BGR2GRAY)
print(rgb_image.min(), rgb_image.max())
print(gray.min(), gray.max())
plt.subplot(1, 2, 1)  # 1 row, 2 columns, position 1
plt.imshow(rgb_image)
plt.title('RGB Image')
plt.axis('off')

plt.subplot(1, 2, 2)  # 1 row, 2 columns, position 2
plt.imshow(gray, cmap='gray')
plt.title('Gray Image')
plt.axis('off')

# Show the images
plt.show()
```

```python
# rotated and scale image
# rotated image means rotated around single point.
image = cv2.imread('img.png')

(h, w) = image.shape[:2]
center = (int(w/2), int(h/2))
# transformation apply on matrix
rotation_matrix = cv2.getRotationMatrix2D(center, 45, 1.0)
rotation_image = cv2.warpAffine(image, rotation_matrix, (w, h))

plt.imshow(rotation_image)
plt.title('Rotated Image')
plt.axis('off')
plt.show()
```

```python
# Draw shape
import numpy as np

image = np.zeros((512, 512, 3), np.uint8)

cv2.rectangle(image,(50, 50), (400, 400), (0, 255, 0), 3) # (start_point, endpoint, color, thickness)
cv2.circle(image, (250,250), 100, (255, 0, 0), 5) # (center, radius, color, thickness)

cv2.line(image, (50, 50), (400, 400), (255, 0, 0), 5) #(start, end, color, thickness)
plt.imshow(image)
plt.title('Original Image')
plt.axis('off')
plt.imshow(image)
```

```python
# image horizontal concat
image1 = cv2.imread('img.png')
image2 = cv2.imread('img_1.png')
print(image1.shape, image2.shape)
height = min(image1.shape[0], image2.shape[0])
width = min(image1.shape[1], image2.shape[1])

image1 = cv2.resize(image1, (width, height))
image2 = cv2.resize(image2, (width, height))
side_by_side_image = cv2.hconcat([image1, image2])
plt.imshow(side_by_side_image)
plt.title('Horizontal Side Image')
plt.axis('off')
plt.show()
```
