# Feature Extraction: Edge Detection Techniques

Edge detection helps us indentify boundaries, shapes, and objects by detecting significant changes in intensity. These are useful in:
- Object detection.
- Image segmentation.
- Shape recognition.
- Medical Imaging(tumour boundary).
- OCR, self-driving, etc.

## Canny Edge Detector:
### Idea:
A multi-stage process:
- Noise reduction (Gaussian Blur).
- Gradient Calculation.
- Non-maximum suppression.
- Hysteresis thresholding (strong and weak edge filter).
### Use Case:
- Detect sharp boundaries like license plates, road lines or edges of a document.

## Sobel Operator
### Idea:
- Uses derivatives to detect edges in horizontal and vertical directions.
- Returns gradient(change in intensity) in x and y directions.
### Use Case:
- Detecting vertical or horizontal object edges -- example; walls, edges of building or barcode patterns.

## Prewitt Operator
### Idea:
- Similar to Sobel but similar kernel(no smoothing effect).
- Less sensitive to noise.
### Use case:
- Basic-edge detection where smoothness is less important. Good for clean, structured shape.

## Laplacian Of Gaussian:
### Idea:
- Combines Gaussian Blur(for noise removal) and laplacian(second derivative).
- Detects edges in all directions and highlights areas of rapid intensity change.
### Use case:
- For precise-edge localization in medical imaging, X-ray, CT scans, etc.

```python
import cv2
from matplotlib import pyplot as plt

image = cv2.imread('img_5.png', cv2.IMREAD_GRAYSCALE)

# Convert to RGB for displaying with matplotlib
# rgb_img = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
kx, ky = cv2.getDerivKernels(1, 0, 3)

canny_img = cv2.Canny(image, 100, 200)
sobel_x = cv2.Sobel(image, cv2.CV_64F, 1, 0, ksize=3)
sobel_y = cv2.Sobel(image, cv2.CV_64F, 0, 1, ksize=3)
sobel_combined = cv2.magnitude(sobel_x, sobel_y)

plt.figure(figsize=(10, 6))
plt.subplot(131), plt.imshow(image, cmap='gray'), plt.title('Original')
plt.subplot(132), plt.imshow(canny_img, cmap='gray'), plt.title('Canny')
plt.subplot(133), plt.imshow(sobel_combined, cmap='gray'), plt.title('Sobel')
plt.tight_layout()
plt.show()
```
