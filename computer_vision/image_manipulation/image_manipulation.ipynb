```python
import cv2
from matplotlib import pyplot as plt

image = cv2.imread('../img.png')
print(image.shape)
# Resize
resize_image = cv2.resize(image, (0, 0), fx=0.5, fy=0.5)
# cropped
cropped_image = image[100:300, 100:300]
#rotated
rotated_image = cv2.rotate(image, cv2.ROTATE_90_CLOCKWISE)

plt.figure(figsize=(12, 8))
plt.subplot(1, 4, 1), plt.imshow(image), plt.title("Original:")  # 1 row, 2 columns, 1st image
plt.subplot(1, 4, 2), plt.imshow(resize_image), plt.title("Resized:")
plt.subplot(1, 4, 3), plt.imshow(cropped_image), plt.title("Cropped:")
plt.subplot(1, 4, 4), plt.imshow(rotated_image), plt.title("Rotated:")
plt.tight_layout()
plt.show()
```

```python
# Color space conversion
import cv2
import matplotlib.pyplot as plt

image = cv2.imread('../img.png')
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
hsv_image = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
lab_image = cv2.cvtColor(image, cv2.COLOR_BGR2LAB)

plt.figure(figsize=(12, 8))

plt.subplot(2, 3, 1), plt.imshow(image), plt.title("Original:")
plt.subplot(2, 3, 2), plt.imshow(gray_image, cmap='gray'), plt.title("Gray:")
plt.subplot(2, 3, 3), plt.imshow(hsv_image), plt.title("HSV:")
plt.subplot(2, 3, 4), plt.imshow(lab_image), plt.title("LAB:")

plt.tight_layout()
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt

image = cv2.imread('../img.png', cv2.IMREAD_GRAYSCALE)
plt.figure(figsize=(12, 8))
plt.title("Grayscale History: ")
plt.xlabel("Pixel value: ")
plt.ylabel("Frequency: ")
hist = cv2.calcHist([image], [0], None, [256], [0, 256])
plt.plot(hist)
plt.show()
```

```python
# apply kernel to image
import cv2
import matplotlib.pyplot as plt
import numpy as np

image = cv2.imread('../img.png', cv2.COLOR_BGR2RGB)
kernal = np.array([
    [0, -1, 0],
    [-1, 5, -1],
    [0, -1, 0]
])
filtered_image = cv2.filter2D(image, -1, kernal)
plt.figure(figsize=(12, 8))
plt.subplot(1,2,1), plt.imshow(image, cmap='gray'), plt.title("Original:")
plt.subplot(1,2,2), plt.imshow(filtered_image), plt.title("Filtered:")
plt.show()
```

```python
# Types of filters
# Gaussian blur
import cv2
import matplotlib.pyplot as plt

image = cv2.imread('../img.png')
if image is None:
    raise ValueError("Image could not be loaded. Please check the file path.")

# Convert BGR to RGB for proper color display
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
gaussian_image = cv2.GaussianBlur(image, (5, 5), 0)

plt.figure(figsize=(12, 8))
plt.subplot(1, 2, 1), plt.imshow(image), plt.title("Original")
plt.axis('off')
plt.subplot(1, 2, 2), plt.imshow(gaussian_image), plt.title("Gaussian Blur")
plt.axis('off')
plt.show()
```

```python
# Edge detection
# Detects vertical and horizontal edges
import cv2
import matplotlib.pyplot as plt

# grayscale cuz intensity doesn't matter for color, so use Grayscale.
image = cv2.imread('../img_2.png', cv2.IMREAD_GRAYSCALE)
# 1, 0 --> vertical edges, 0, --> horizontal edges
# cv2.CV_64F --> avoid loss of negative gradients.
# ksize --> kernel size
sobel_x = cv2.Sobel(image, cv2.CV_64F, 1, 0, ksize=5)
sobel_y = cv2.Sobel(image, cv2.CV_64F, 0, 1, ksize=5)

plt.figure(figsize=(12, 8))
plt.subplot(1, 3, 1), plt.imshow(image, cmap='gray'), plt.title("Original")
plt.axis('off')
plt.subplot(1, 3, 2), plt.imshow(sobel_x, cmap='gray'), plt.title("Sobel X")
plt.axis('off')
plt.subplot(1, 3, 3), plt.imshow(sobel_y, cmap='gray'), plt.title("Sobel Y")
plt.axis('off')
plt.show()

```
