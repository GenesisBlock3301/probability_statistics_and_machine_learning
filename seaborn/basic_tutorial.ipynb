# Seaborn Plot Type
- Uni-variant -- x only(contain only one axis information)
- Bivariant --> x and y only(contain only two axis information)
- Tri-variant --> x,y,z (contains three axis information)

Commonly used plot types Seaborn use:
- **Scatter plot:** A scatter plot visualize the relationship between two variables.
- **Line Plot:** A line plot used to visualize the trend of a variable over a time.
- **Histogram:** A histogram is used to visualize the distribution of variables.
- **Box plot:**  A box plot is used to visualize the distribution of a variables.
- **Violin:** This is similar to box plot but provides more detail view of the distribution of the data.
- **Heatmap:** This is used to visualize the correlation between multiple variables.
- **Pair Plot:** This is used to visualize the relationship between variables.

```python
#Scatter plot
fmri = sns.load_dataset("fmri")
tips = sns.load_dataset("tips")
fmri.head(10)
```

```python
sns.scatterplot(x="timepoint", y="signal", data=fmri.head(10))
plt.show()
```

```python
# hue for differentiate eac data point different color
sns.scatterplot(x="timepoint", y="signal",hue="region",
                style="event",
                data=fmri.head(5))
plt.show()
```

```python
tips.head()
```

```python
sns.scatterplot(x="day",y="tip", data=tips, hue="day")
plt.show()
```
