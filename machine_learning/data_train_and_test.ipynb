```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

np.random.seed(2)

x = np.random.normal(3, 1, 100)
y = np.random.normal(150, 40, 100) / x
x, y
```

```python
plt.scatter(x, y)
plt.show()
```

```python
train_x = x[:80]
train_y = y[:80]

test_x = x[80:]
test_y = y[80:]
```

```python
# display the training set
plt.scatter(train_x, train_y)
plt.show()
```

```python
# display the test dat
plt.scatter(test_x, test_y)
plt.show()
```

```python
poly_fit = np.polyfit(train_x, train_y, 3)
print("polynomial fit:", poly_fit)
my_model = np.poly1d(poly_fit)
print("my_model", my_model)
my_line = np.linspace(0, 6, 100)
print("my_line", my_line)
plt.scatter(train_x, train_y)
model_data = my_model(my_line)
print("model_data", model_data)
plt.plot(my_line, model_data)
plt.show()
```

```python
from sklearn.metrics import r2_score

r2 = r2_score(train_y, my_model(train_x))
print('r2 score:', r2)
```
