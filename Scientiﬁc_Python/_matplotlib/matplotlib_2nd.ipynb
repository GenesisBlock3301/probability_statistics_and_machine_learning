#### Line Style

```python
# Line style
# dotted, dashed
import matplotlib.pyplot as plt
import numpy as np
y_points = np.array([3,10,4,5,6])
plt.plot(y_points, linestyle='dashed')
plt.show()
```

#### Line color

```python
import matplotlib.pyplot as plt
import numpy as np
y_points = np.array([3,10,4,5,6])
# line color can be red, yellow, hotpink etc
plt.plot(y_points, ls='-.', color='green')
plt.show()
```

#### Line width

```python
import matplotlib.pyplot as plt
import numpy as np
y_points = np.array([3,10,4,5,6])
# color, line width
plt.plot(y_points, ls='-.', color='r',linewidth='20.5')
plt.show()
```

#### Multiple Lines

```python
import matplotlib.pyplot as plt
import numpy as np

y1 = np.array([3, 8, 1, 10])
y2 = np.array([6, 2, 7, 11])

plt.plot(y1, color="green", ls="-.")
plt.plot(y2, color="red", ls=":")

plt.show()
```
