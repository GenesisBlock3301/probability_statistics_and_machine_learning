## Why do we need matplotlib?
Matplotlib is a plotting library for Python that is used for data
visualization. It provides a wide range of tools and features for
creating various types of plots, including line plots, bar plots,
scatter plots, histograms, and many others.

Matplotlib was created by John D. Hunter.

List of topics:
- Pyplot
- Plotting
- Markers

```python
import matplotlib.pyplot as plt
plt.plot(range(10))
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x_points = np.array([0,6])
y_points = np.array([0,250])

plt.plot(x_points, y_points)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x_points = np.array([1,8])
y_points = np.array([3,10])

plt.plot(x_points, y_points)
plt.show()
```

```python
# Draw points instead of line
import matplotlib.pyplot as plt
import numpy as np

x_points = np.array([1,8,6,4,6])
y_points = np.array([3,10,4,5,6])

plt.plot(x_points, y_points, 'o')
plt.show()
```

```python
# Default X-points 0 1 2 3 etc
import matplotlib.pyplot as plt
import numpy as np

# x_points = np.array([1,8,6,4,6])
y_points = np.array([3,10,4,5,6])

plt.plot(y_points, 'o')
plt.show()
```

```python
# You can use the keyword argument marker to emphasize each point with a specified marker:

import matplotlib.pyplot as plt
import numpy as np

# x_points = np.array([1,8,6,4,6])
y_points = np.array([3,10,4,5,6])
# 'g' green color
# ms size of points
# mec edge color
plt.plot(y_points, 'o:g', ms=10, mec='r')
plt.show()
```
