- Create level and title for a plot
- Adding Grid line

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

# front properties
font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}

plt.title("Title", fontdict=font1, loc='left')
plt.xlabel("X-level", fontdict=font2)
plt.ylabel("Y-level", fontdict=font2)
plt.grid(axis='x') #color = 'green', linestyle = '--', linewidth = 0.5
plt.plot(x,y)
plt.show()
```

### Multiple plot in one figure

```python
import numpy as np

#plot-1
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 40])
#the figure has 1 row, 2 columns, and this plot is the first plot.
plt.subplot(1,2,1)
plt.title("First plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.plot(x,y)

# plot-2
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(1, 2, 2)
plt.title("Second plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.plot(x,y)

plt.suptitle("Example of PLOT")
plt.show()
```

```python
import numpy as np

#plot-1
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 40])
#the figure has 2 row, 1 columns, and this plot is the first plot.
plt.subplot(2,1,1)
plt.plot(x,y)

# plot-2
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])
# 2 row, 1 column, 2nd plot.
plt.subplot(2, 1, 2)
plt.plot(x,y)
```
