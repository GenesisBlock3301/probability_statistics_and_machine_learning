#### Creating Bar & Histogram
- color
- width

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])
# colors = np.random.randint(4, size=10)
plt.bar(x,y, color='red',width=0.1)
plt.show()
```

#### Horizontal Bars

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])
# barh() for showing horizontal
plt.barh(x,y)
plt.show()
```
