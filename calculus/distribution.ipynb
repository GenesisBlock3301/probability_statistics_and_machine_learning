# Uniform distribution-

üìñ Story:
In a school in Dhaka, the teacher brings a jar with 5 different colored candies üç¨ ‚Äî red, green, blue, yellow, and orange.
She tells the students:

    "Each of you will pick one candy blindfolded. Every candy has the same chance of being picked."

This is a Uniform Distribution ‚Äî because all outcomes are equally likely.

üìä Characteristics:

    All candies: equal probability.

    This is like drawing a fair lottery.

üßÆ Probability:
P(any color)=1/5=0.20


```python
import numpy as np
import matplotlib.pyplot as plt

colors = ['Red', 'Yellow', 'Green', 'Blue', 'Magenta']
picks = np.random.choice(colors, size=1000, replace=True)

plt.hist(picks, bins=len(colors), color='skyblue', edgecolor='black')
plt.title('Uniform Distribution')
plt.xlabel('Color')
plt.ylabel('Count')
plt.show()
```

üîî 2. Normal Distribution ‚Äî üìö Exam Scores in a Village School
üìñ Story:

In a school in Rajshahi, 100 students took a Math exam.

    Most students scored around 50‚Äì60

    A few very smart students got 90+

    A few struggling ones got below 30

This forms a bell-shaped curve ‚Äî Normal Distribution.
üìä Characteristics:

    Most students are average

    Very high or low scores are rare

    Looks like a hill (bell curve)

üéØ Real-Life Insight:

This is true for many things in life:

    People‚Äôs height in Bangladesh üßç‚Äç‚ôÇÔ∏è

    Daily working hours for a jobholder üïí

    Prices of same vegetables in different bazaars ü•¶



```python
import numpy as np
import matplotlib.pyplot as plt

# Simulate exam scores with mean=60, std=10
scores = np.random.normal(loc=60, scale=10, size=1000)

plt.hist(scores, bins=30, color='orange', edgecolor='black')
plt.title("Normal Distribution of Math Scores")
plt.xlabel("Score")
plt.ylabel("Number of Students")
plt.axvline(x=60, color='red', linestyle='--', label='Average (60)')
plt.legend()
plt.show()
```
