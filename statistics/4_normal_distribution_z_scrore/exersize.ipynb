```python
low, high = df.price_per_sqft.quantile([0.001,0.999])
low, high
```

```python
outliers = df[(df.price_per_sqft < low) | (df.price_per_sqft > high)]
outliers
# outliers.shape
# """here we can see 1365 is too lower price per sqft"""
```

```python
df2 = df[(df.price_per_sqft > low) & (df.price_per_sqft < high)]
df2
```

```python
# In above outliers 28 outliers reduce
```

```python
max_limit = df2.price_per_sqft.mean() + 4*df2.price_per_sqft.std()
low_limit = df2.price_per_sqft.mean() - 4*df2.price_per_sqft.std()
low_limit, max_limit
```

```python
remove_count = df2[(df2.price_per_sqft < low_limit) | (df2.price_per_sqft > max_limit)]
remove_count.shape, df2.shape
```

```python
df3 = df2[(df2.price_per_sqft > low_limit) & (df2.price_per_sqft < max_limit)]
df3.shape
```

```python
number_of_outlier_remove = df2.shape[0] - df3.shape[0]
number_of_outlier_remove
```

```python
plt.hist(df3.price_per_sqft, bins=20, rwidth=0.8)
plt.xlabel("Price per square ft")
plt.ylabel("Count")
plt.show()
```
