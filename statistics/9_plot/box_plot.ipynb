```python
import seaborn as sns
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
iris = sns.load_dataset("iris")
iris
```

```python
sns.boxenplot(x="species", y="petal_length", data=iris)
plt.show()
```

```python
sns.violinplot(x="species", y="petal_length", data=iris, size=8)
plt.show()
```
