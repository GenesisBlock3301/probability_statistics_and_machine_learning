```python
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
```

```python
iris = sns.load_dataset("iris")
iris.head()
```

```python
iris.shape, iris.columns
```

```python
# If all values are same count, in that case this is called balanced dataset.
# if not this is called im-balanced dataset.
iris["species"].value_counts()
```

```python
sns.set_style("whitegrid")
sns.FacetGrid(iris, hue="species", height=4)\
    .map(plt.scatter, "sepal_length", "sepal_width")\
    .add_legend();
plt.show()
```

```python
sns.pairplot(iris, hue="species", height=4 )
plt.show()
```

```python
#Histogram
iris = load_iris()
iris_setosa = iris.data[iris.target == 0]
iris_virginica = iris.data[iris.target == 1]
iris_versicolor = iris.data[iris.target == 2]
```

```python
sns.set_style("whitegrid")
sns.histplot(iris_setosa[:,2], kde=True, label="Setosa", color="blue", multiple="layer")
sns.histplot(iris_versicolor[:,2],kde=True, label="Versicolor", color="green", multiple="layer")
sns.histplot(iris_virginica[:,2], kde=True, label="Virginica ", color="red", multiple="layer")
# Add labels and legend
plt.xlabel("Petal Length")
plt.ylabel("Frequency")
plt.title("Histogram of Petal Length for Different Iris Species")
plt.legend()
plt.show()
```
